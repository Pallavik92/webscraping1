```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.moneycontrol.com/stocks/marketinfo/marketcap/bse/index.html")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
table1=soup.find_all('div',class_="undefined undefined undefined")
table1
```

```python
all_data=[]
for i in table1:
  for j in i.find_all("td"):
    all_data.append(j.text.replace("\n"," "))
all_data
```

```python
company_name=all_data[0::6]
company_name
```

```python
last_price=all_data[1::6]
last_price
```

```python
percent_change=all_data[2::6]
percent_change
```

```python
wk_52_high=all_data[3::6]
wk_52_high
```

```python
wk_52_low=all_data[4::6]
wk_52_low
```

```python
market_cap=all_data[5::6]
market_cap
```

```python
import pandas as pd
top_company=pd.DataFrame({})
top_company["COMPANY NAME"]=company_name
top_company["LAST PRICE"]=last_price
top_company["%CHANGE"]=percent_change
top_company["52WK HIGH"]=wk_52_high
top_company["52WK LOW"]=wk_52_low
top_company["MARKET CAP"]=market_cap
```

```python
top_company
```

```python
top_company.to_csv("money control")
```
