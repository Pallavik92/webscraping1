```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://fastestlaps.com/lists/top-most-expensive-cars")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
table=soup.find_all("table",class_="table table-striped")
table
```

```python
all_data=[]
for i in table:
    for j in i.find_all("td"):
      all_data.append(j.text.replace("\n",""))
all_data
```

```python
rank=all_data[0::3]
rank
```

```python
make_model=all_data[1::3]
make_model
```

```python
price=all_data[2::3]
price
```

```python
import pandas as pd
```

```python
top_car=pd.DataFrame({})
top_car["carRANK"]=rank
top_car["MAKE MODEL"]=make_model
top_car["PRICE"]=price
```

```python
top_car
```

```python
top_car.to_csv("car_models")
```
